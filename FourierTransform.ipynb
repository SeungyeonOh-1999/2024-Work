# 1. Fourier Transform 

### 1. Toy Example

$\omega$ : Angular Frequency  
$f$ : Frequency   
$p$ : Period   
$\omega = 2 \pi f$  

<br>
The original signal consists of a cosine wave with a frequency of 5Hz and an amplitude of 1, and a sine wave with a frequency of 10Hz and an amplitude of 3. Perform a Fourier transform on the signal measured over a duration of 10 seconds.

```python
import numpy as np
import matplotlib.pyplot as plt


fs = 100     # fs: number of samples per second   
t = np.linspace(0, 10, fs * 100, endpoint=False) 

original_signal = 1 * np.cos(2 * np.pi * 5 * t) + 3 * np.sin(2 * np.pi * 10 * t)
```

```python
signal_5Hz = np.cos(2 * np.pi * 5 * t) 
signal_10Hz = 3 * np.sin(2 * np.pi * 10 * t )

plt.figure(figsize= (30, 20))

plt.subplot(3, 1, 1)
plt.plot(t, signal_5Hz)
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 2)
plt.plot(t, signal_10Hz)
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.subplot(3, 1, 3)
plt.plot(t, original_signal)
plt.title('Combined Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.show()
```

```python
fft = np.fft.fft(original_signal) / len(t)
fft_freq = np.fft.fftfreq(len(t), 1/fs)     
fft_magnitude = np.abs(fft)

# "np.fft.fftfreq" returns periodic components from the result of Fourier transform. 
# The fist half of "fft_freq" contains positive values, while the rest contains negative values. 
# Becasue Fast Fourier Transform (FFT) results increase proportionally with the number of samples,
# we need to normalize them through dividing by the length of the original signal.

plt.figure(figsize=(30, 12))

plt.subplot(2, 1, 1)
plt.plot(t, original_signal)
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(fft_freq, fft_magnitude)
plt.title('Fourier Transform Result')
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.grid()
plt.show()
```
